## 3.1.4 Renaming columns

```r
# rename 'price' to 'price_dollar'
diamonds <- rename(diamonds, price_dollar = price)
# preview of the modified dataset with different column name
head(diamonds)
```

## 3.1.5 Adding/changing columns

```r
# add new column with price in chinese currency
diamonds <- mutate(diamonds, price_yuan = price_dollar*7.2)
head(diamonds)
```

```r
# show the descriptive data of column 'price_yuan'
summary(diamonds$price_yuan)
```

```r
# change values in 'price_yuan' column to read "low", "rather low", "rather high", and "high" instead of numbers
diamonds <- mutate(diamonds, price_yuan = ifelse(price_yuan <= 3000, 'low',
                                          ifelse(price_yuan > 3000 & price_yuan <= 6000, 'rather low',
                                          ifelse(price_yuan > 6000 & price_yuan <= 9000, 'rather high',
                                          ifelse(price_yuan >9000, 'high', NA)))))
head(diamonds)
```

## 3.1.6 Filtering data

```r
# filter data from 'rather high' objects in price_yuan only
diamond_price_ratherhigh <- filter(diamonds, price_yuan == 'rather high')
```

## 3.1.7 Piping

```r
# take diomands data, filter for 'rather high', drop the column 'price_yuan', and add a new column calculating price in euro
diamonds <- diamonds %>%
  filter(price_yuan == 'rather high') %>%
  select(-price_yuan) %>%
  mutate(price_euro = price_yuan*0.14)
```

## 3.1.8 Aggregating data

```r
head(diamonds)
```

```r
# compare price averages across color, use color as the grouping variable and calculate means for the other variables
class(diamonds$color)
diamonds_byColor <- diamonds %>%
  group_by(color) %>%
  summarize(avg_price_dollar = mean(price_dollar),
            avg_depth = mean(depth))
class(diamonds_byColor)
diamonds_byColor
```

## 3.2.1 Histograms

```r
# create a histogram of the variable 'depth'
ggplot(data = diamonds, mapping = aes(x = depth)) + geom_histogram(bins = 100)
```

```r
# want relative frequency on the y-axis (a density plot)
ggplot(data = diamonds, mapping = aes(x = depth)) + geom_histogram(bins = 100, aes(y = ..density..))
```

```r
# adding axis labels and changing the colors and theme
ggplot(data = diamonds, mapping = aes(x = depth)) + geom_histogram(bins = 100, aes(y = ..density..), fill = 'lightgreen') + xlab('diamond depth information') + ggtitle('distribution of diamond depth information') + theme_light()
```

## 3.2.2 Scatterplots

```r
# create a scatterplot of carat on price_dollar
ggplot(data = diamonds, mapping = aes(x = carat, y = price_dollar)) + geom_point() + theme_light()
```

```r
# distinguish by different color
ggplot(data = diamonds, mapping = aes(x = carat, y = price_dollar, color = color)) + geom_point() + theme_light()
```

## 3.2.3 Box & Whisker Plots

```r
# create a box and whisker plot of carat (continuous) on color (categorical)
ggplot(data = diamonds, mapping = aes(x = carat, y = color, color = color)) + geom_boxplot() + xlab('diamond carat') + ylab('diamond color') + theme_light()
```

# Chapter 4 Basic Data Analysis

## 4.1 The Distribution of Data

```r
# summarize PropFemaleTopQuartile
colnames(paygap)
```

```r
# create a histogram of the variable FemaleLowerQuartile
ggplot(aes(x = FemaleLowerQuartile), data = paygap) + 
  geom_histogram(bins = 50, aes(y = ..density..)) +
  xlab('proportion of females in lower-earning quartile') +
  theme_bw()
```

```r
# summary statistics of 'FemaleLowerQuartile'
summary(paygap$FemaleLowerQuartile)
```

```r
# determine the mean of 'FemaleLowerQuartile'
mean(paygap$FemaleLowerQuartile)
# determine the median of 'FemaleLowerQuartile'
median(paygap$FemaleLowerQuartile)
# determine the SD of 'FemaleLowerQuartile'
sd(paygap$FemaleLowerQuartile)
# determine the Median Absolute Deviation (MAD) of 'FemaleLowerQuartile'
mad(paygap$FemaleLowerQuartile)
# determine the confidence intervals of 'DiffMeanHourlyPercent'
confidence_interval(data = paygap$DiffMeanHourlyPercent, conflevel = 0.95)
```

## 4.3.1 One-sample t-test for a mean

```r
# run one sample t-test on DiffMeanHourlyPercent
t.test(paygap$DiffMeanHourlyPercent, mu = 0)
```

## 4.3.3 Two-sample t-test for a difference in means

```r
# determine the mean difference of FemaleBonusPercent and MaleBonusPercent
t.test(paygap$FemaleBonusPercent, paygap$MaleBonusPercent)
# p < 0.05, there is a difference between these two variables
```
